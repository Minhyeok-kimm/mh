# 기말과제

## spaceship-titanic
- PassengerId:각 승객의 고유 ID로 gggg_pp 형식으로 여기서 gggg는 승객이 함께 여행하는 그룹을 나타내고 pp는 그룹 내 번호이며, 그룹의 사람들은 종종 가족 구성원이지만 항상 그런 것은 아님
- HomePlanet: 승객이 출발한 행성
- CryoSleep: 승객이 항해 기간 동안 수면을 일시 중단하도록 선택했는지 여부
- Cabin: 승객이 머물고 있는 객실 번호로 Deck/num/side 형식을 취하며, 여기서 side는 Port의 경우 P 또는 Starboard의 경우 S일 수 있음
- Destination: 승객이 내릴 행성
- Age: 승객의 나이
- VIP: 승객이 항해 중 특별 VIP 서비스를 지불했는지 여부
- RoomService, FoodCourt, ShoppingMall, Spa, VRDeck: 우주선 타이타닉의 다양한 고급 편의 시설에 대해 승객이 청구한 금액
- Name: 승객의 성과 이름
- Transported: 승객이 다른 차원으로 운송되었는지 여부로 Target Feature

## 4. 결측치 처리

```python
spaceship.isnull().sum()
```

```python
spaceship.dropna(how = 'all') # 전체 값이 nan인 경우 삭제
```

```python
spaceship['HomePlanet'].value_counts()
```

```python
# 'HomePlanet' 열 결측치에 'Earth' 채우기
spaceship['HomePlanet'].fillna('Earth', inplace = True)
spaceship['HomePlanet'].value_counts()
```

```python
spaceship['CryoSleep'].value_counts()
```

```python
# 'CryoSleep' 열 결측치에 False 채우기
spaceship['CryoSleep'].fillna(False, inplace = True)
spaceship['CryoSleep'].value_counts()
```

```python
spaceship['Cabin'].value_counts()
```

```python
# 'Cabin' 열 결측치에 'G/734/S' 채우기
spaceship['Cabin'].fillna('G/734/S', inplace = True)
spaceship['Cabin'].value_counts()
```

```python
spaceship['Destination'].value_counts()
```

```python
# 'Destination' 열 결측치에 'TRAPPIST-1e' 채우기
spaceship['Destination'].fillna('TRAPPIST-1e', inplace = True)
spaceship['Destination'].value_counts()
```

```python
spaceship[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].describe()
```

```python
# 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck' 열 결측치에 각각의 mean 값 채우기
spaceship['RoomService'].fillna(spaceship['RoomService'].mean(), inplace = True)
spaceship['FoodCourt'].fillna(spaceship['FoodCourt'].mean(), inplace = True)
spaceship['ShoppingMall'].fillna(spaceship['ShoppingMall'].mean(), inplace = True)
spaceship['Spa'].fillna(spaceship['Spa'].mean(), inplace = True)
spaceship['VRDeck'].fillna(spaceship['VRDeck'].mean(), inplace = True)
spaceship[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].describe() # Count 개수 열 개수와 동일 확인
```

```python
spaceship['Age'].describe()
```

```python
# 'Age' 열 결측치에 mean 값 채우기
spaceship['Age'].fillna(spaceship['Age'].mean(), inplace = True)
spaceship['Age'].value_counts()
```

```python
spaceship['VIP'].value_counts()
```

```python
# 'VIP' 열 결측치에 False 채우기
spaceship['VIP'].fillna(False, inplace = True)
spaceship['VIP'].value_counts()
```

```python
spaceship['Name'].value_counts()
```

```python
# 'Name' 열 결측치에 'Unnown' 채우기
spaceship['Name'].fillna('Unnown', inplace = True)
spaceship['Name'].value_counts()
```

## 10. apply, lambda로 데이터 가공

```python
#' 'Child_Adult' 열을 생성하고 Child를 넣어주고, 만약 20세 이상이면 Adult 문자열 삽입
spaceship['Child_Adult'] = spaceship['Age'].apply(lambda x: 'Child' if x <= 20 else 'Adult' )
spaceship[['Age','Child_Adult']].head(10)
```

```python
# 'Age_Cat' 열을 생성하고 15세 미만은 "Child", 15세 이상 20세 미만은 "Youth", 20세 이상은 "adult",  65세 이상은 "Elderly" 채우기
spaceship['Age_cat'] = spaceship['Age'].apply(lambda x: 'Child' if x < 15 else ('Youth' if 15 <= x < 20 \
                                                                                else ('Adult' if x <= 65 else 'Elderly')))
spaceship['Age_cat'].head(15)
```

```python
# value_counts()로 갯수 확인
spaceship['Age_cat'].value_counts()
```
